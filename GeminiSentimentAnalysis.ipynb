```python
import pandas as pd

# Assuming 'df' is your DataFrame and 'commentText' is the column with text

# Function to perform sentiment analysis on a batch of comments
def analyze_sentiment_batch(comments_batch):
    try:
        # Join the comments into a single string separated by a delimiter
        batched_text = ' || '.join(comments_batch)
        # Send the batched text to the API
        response = model.generate_content("Analyze the sentiment of this text with [one word] English sentiment of varying emotion, not only negative or positive or mixed: " + batched_text)
        # Split the response back into individual sentiments
        sentiments = response.text.split(' || ')
        return sentiments
    except ValueError as e:
        # If there's a ValueError, it means the sentiment was blocked
        print("Sentiment analysis blocked:", e)
        # Return a list of 'blocked' equal to the number of comments in the batch
        return ['blocked'] * len(comments_batch)

# Function to process the entire DataFrame and add sentiments as a new column
def add_sentiment_column(df, batch_size=1):
    # Create an empty list to store sentiment results
    sentiment_results = []

    # Convert the 'commentText' column to a list of comments
    comments = df['commentText'].tolist()

    # Process comments in batches
    for i in range(0, len(comments), batch_size):
        # Create a batch of comments
        batch = comments[i:i + batch_size]
        # Analyze sentiment for the current batch
        sentiments = analyze_sentiment_batch(batch)
        # Extend the sentiment results list
        sentiment_results.extend(sentiments)

    # Ensure the list is the same length as the DataFrame
    sentiment_results += [''] * (len(df) - len(sentiment_results))

    # Add the sentiments as a new column to the DataFrame
    df['Sentiment'] = sentiment_results

# Example usage:
add_sentiment_column(df)
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and it has a 'Sentiment' column with string values

# Count the occurrences of each unique sentiment value
sentiment_counts = df['Sentiment'].value_counts()

# Prepare data for the pie chart
labels = sentiment_counts.index.tolist()
sizes = sentiment_counts.values.tolist()

# Plot the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add a title
plt.title('Sentiment Distribution')

# Show the plot
plt.show()
```
